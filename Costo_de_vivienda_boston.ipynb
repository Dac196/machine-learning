DataSet costo de vivienda en boston

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from numpy import genfromtxt
import pandas as pd
dataset = pd.read_csv('https://raw.githubusercontent.com/m-mehdi/tutorials/main/boston_housing.csv', delimiter=',',header=None)
X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

Se importa el conjunto de datos y se crean 2 arreglos, uno con todos los datos excepto la ultima columna que corresponde al valor calculado de la vivienda el segundo solamente guarda el valor de la vivienda. Este sera el que se utilice para comparar el nivel de acierto del script

Se revisa la información general del data set en este caso todos los datos son númericos no hay variables categoricas por lo que se puede continuar con el proceso de entrenamiento del modelo sin la necesidad de excluir columnas o convertir variables catregoricas

```python
X
```

Se crean los conjuntos de entrenamientos y de testing

En este ejercicio se separa un 30% del dataset para utilizarlo en el entrenamiento del modelo.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=0)
```

Se realiza una regresión líneal y se entrena el modelo. Despues se pregunta por los porcentajes de acierto que tuvo en los 2 conjuntos es decir el de entrenamiento y el de testing.

Se puede evidenciar que bajo estas caracteristicas el porcentaje de acierto en el modelo de entrenamiento fue muy alto sin embargo en el de testing el porcenteja es considerablemente más bajo, lo que indica que probablemente se presento un overfitting  y aprendio sobre aprendio del ruido del conjunto de entrenamiento y no es un modelo optimo para predecir resultados con nuevos modelos.

```python
lr = LinearRegression().fit(X_train, y_train)

print(f"Porcentaje de acierto con el modelo de entrenamiento: {lr.score(X_train, y_train):.2f}")
print(f"Porcentaje de acierto con el modelo de testing: {lr.score(X_test, y_test):.2f}")
```

Se utiliza la regularizacion L2 o Ridge con los mismos conjuntos de entrenamiento, se realizan diferentes pruebas de valor alpha, encontrando que con un alpha = 0.1 se mantiene un porcentaje alto en el conjunto de entranmiento y el valor del conjunto de testing tambien incremente su porcentaje de acierto sin embargo la diferencia de aciertos entre los 2 conjuntos de datos sigue siendo alto.

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=0.1).fit(X_train, y_train)
print(f"Porcentaje de acierto con el modelo de entrenamiento regularizacion Ridge: {ridge.score(X_train, y_train):.2f}")
print(f"Porcentaje de acierto con el modelo de testing regularizacion Ridge: {ridge.score(X_test, y_test):.2f}")
```

Se utiliza la regularización L1 o Lasso con los conjuntos de entrenamiento y un valor de alpha = 0.05. Con estas condiciones se encuentra el porcentaje de acierto en el conjunto de testing incrementa sin embargo el porcentaje de acierto del conjunto de entrenamiento disminuye.

```python
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=0.05).fit(X_train, y_train)
print(f"Porcentaje de acierto con el modelo de entrenamiento regularizacion Lasso: {lasso.score(X_train, y_train):.2f}")
print(f"Porcentaje de acierto con el modelo de testing regularizacion Lasso: {lasso.score(X_test, y_test):.2f}")
```

Teniendo en cuenta que Lasso disminuye el peso de algunas variables a 0 se consulta cuantas variables esta utilizando para la predicción del conjunto de datos. En este caso con el valor de alpha asignado se estan utilizando 18 variables de las 103 que tiene el modelo.

Sin embargo si se aumenta el valor de alpha para que incluya más variables los porcentajes de acierto no aumentan significativamente y no ayuda a optimizar el modelo.

```python
print(f"Number of features: {sum(lasso.coef_ != 0)}")
```
