```python
import pandas as pd
Null_values =['0'] # se identifican valores nulos dentro del data set
nRowsRead=None
COLUMNS = ['age','workclass','fnlwgt','education','education_num',
           'marital','ocupation','relationship','race','sex','capital_gain',
           'capital_loss', 'hours_week','native_country','label']
PATH ='adult.data'
df_adult = pd.read_csv(PATH, delimiter=',', nrows = nRowsRead, names = COLUMNS, na_values = Null_values)
df_adult.dataframeName ='adult.csv'
nRow,nCol = df_adult.shape
print(df_adult)
```
